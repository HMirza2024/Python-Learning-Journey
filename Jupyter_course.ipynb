```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Generate a random 2D array of integers
random_array = np.random.randint(0, 100, size=(5, 10))

# Compute mean, standard deviation, and variance along the second axis
mean_values = np.mean(random_array, axis=1)
std_deviation = np.std(random_array, axis=1)
variance = np.var(random_array, axis=1)

print("Mean along second axis:", mean_values)
print("Standard deviation along second axis:", std_deviation)
print("Variance along second axis:", variance)

# Load Iris dataset
iris = load_iris()
classes, counts = np.unique(iris.target, return_counts=True)

# Plot bar chart
plt.bar(classes, counts, tick_label=iris.target_names)
plt.xlabel('Class Label')
plt.ylabel('Frequency')
plt.title('Frequency of each class label in Iris dataset')
plt.legend(['Setosa', 'Versicolor', 'Virginica'])
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Load the iris dataset
df = pd.read_csv("iris.csv")
# Print the first 5 rows of the dataset
print(df.head)
print(df.columns)
# a. Plot bar chart to show the frequency of each class label in the data
# Count the frequency of each iris class label
class_counts = df['variety'].value_counts()
plt.figure(figsize=(8, 6))
class_counts.plot(kind='bar', color='skyblue')
plt.title('Frequency of Each Iris Class Label')
plt.xlabel('Class Label')
plt.ylabel('Frequency')
plt.show()

```

```python
# b. Draw a scatter plot for Petal width vs Sepal width and fit a regression line
sns.regplot(x='sepal width (cm)', y='petal width (cm)', data=df)
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Petal Width (cm)')
plt.title('Scatter Plot: Petal Width vs Sepal Width')
plt.show()

```

### Code Explanation

```python
# c. Plot density distribution for feature petal length
sns.kdeplot(data=df, x='petal length (cm)', hue='variety', fill=True)
plt.xlabel('Petal Length (cm)')
plt.ylabel('Density')
plt.title('Density Distribution: Petal Length')
plt.show()

```

```python
# d. Use a pair plot to show pairwise bivariate distribution in the Iris Dataset
sns.pairplot(df, hue='variety')
plt.show()


```

```python
# e. Draw heatmap for the four numeric attributes (so remove extra attribute)
numeric_attributes = df.drop('variety', axis=1)
sns.heatmap(numeric_attributes.corr(), annot=True, cmap='coolwarm')
plt.title('Heatmap: Correlation of Numeric Attributes')
plt.show()
```

```python
import numpy as np

# Given 1D array
arr = np.array([1, 0, 5, 0, np.nan, 3, np.nan])

# Test for zero elements
zero_indices = np.where(arr == 0)[0]

# Test for non-zero elements
non_zero_indices = np.where(arr != 0)[0]

# Test for NaN elements
nan_indices = np.where(np.isnan(arr))[0]

print("Indices of zero elements:", zero_indices)
print("Indices of non-zero elements:", non_zero_indices)
print("Indices of NaN elements:", nan_indices)
```

```python
import numpy as np

# Create random arrays Array1, Array2, and Array3 of the same size
size = (3, 4)  # Example size
Array1 = np.random.rand(*size)
Array2 = np.random.rand(*size)
Array3 = np.random.rand(*size)

# Subtract Array2 from Array3 and store in Array4
Array4 = Array3 - Array2

# Create Array5 having two times the values in Array1
Array5 = 2 * Array1

# Find covariance of Array1 with Array4
covariance_array1_array4 = np.cov(Array1.flatten(), Array4.flatten())[0, 1]

# Find correlation of Array1 with Array5
correlation_array1_array5 = np.corrcoef(Array1.flatten(), Array5.flatten())[0, 1]

print("Array1:")
print(Array1)
print("\nArray2:")
print(Array2)
print("\nArray3:")
print(Array3)
print("\nArray4 (Array3 - Array2):")
print(Array4)
print("\nArray5 (2 * Array1):")
print(Array5)
print("\nCovariance of Array1 with Array4:", covariance_array1_array4)
print("Correlation of Array1 with Array5:", correlation_array1_array5)
```
